# Demo: Export icechunk raster data to geotiff

Raster topobathymetry data made into icechunk stores is not inherently visibile with traditional GIS methods. Use this method to retrieve a dataset and export to geotiff. 

Requires:
- `.env` with AWS credentials

```python
from icefabric_tools.icechunk import IcechunkS3Repo, NGWPCLocations
from dotenv import load_dotenv
```

```python
# before running this cell, make sure you created a .env file in this directory with your AWS credentials in it
load_dotenv()  
```

Open the dataset to check information and variable name.

```python
repo = IcechunkS3Repo(location=NGWPCLocations.TOPO_HA_30M_IC.path)
ds = repo.retrieve_dataset()
ds
```

Save the full dataset to tiff.

```python
var_name = "elevation"
dest = "./hawaii_topo_30.tif"
repo.retrieve_and_convert_to_tif(dest=dest, var_name=var_name)
```

Save a clipped bounding box of dataset.

```python
var_name = "elevation"
dest = "./hawaii_clipped_topo_30.tif"
minx, miny, maxx, maxy = -159, 21, -158, 22
repo.retrieve_and_convert_to_tif(dest=dest, var_name=var_name, minx=minx, miny=miny, maxx=maxx, maxy=maxy)
```
